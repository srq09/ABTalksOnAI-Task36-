```python
from openai import OpenAI

# YAHAN APNI OPENAI API KEY PASTE KAREIN
API_KEY = "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"

if API_KEY == "PASTE_YOUR_NEW_OPENAI_API_KEY_HERE" or not API_KEY.strip():
    raise ValueError("Please paste your OpenAI API key inside API_KEY variable.")

client = OpenAI(api_key=API_KEY)


weekly_review_questions = [
    {
        "category": "Prompt Engineering",
        "question": "Explain what makes a good AI prompt."
    },
    {
        "category": "LLM Workflow",
        "question": "Describe the workflow of an LLM-based chatbot."
    },
    {
        "category": "Tool Integration",
        "question": "How can external tools improve AI applications?"
    },
    {
        "category": "Memory",
        "question": "Why is conversational memory important in chatbots?"
    },
    {
        "category": "Evaluation",
        "question": "How do you evaluate GenAI responses?"
    }
]


class GenAIEvaluator:
    def __init__(self):
        self.total_score = 0
        self.results = []

    def evaluate_response(self, category, user_answer):
        prompt = f"""
Evaluate this answer professionally.

Category: {category}
Answer: {user_answer}

Score out of 10 based on:
1. Accuracy
2. Clarity
3. Completeness
4. Practical Understanding

Return EXACT format:
Score: X/10
Feedback: detailed explanation
"""

        try:
            response = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[
                    {"role": "system", "content": "You are a professional GenAI evaluator."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.3
            )

            result = response.choices[0].message.content

            try:
                score_line = result.split("\n")[0]
                score = int(score_line.split(":")[1].split("/")[0].strip())
                self.total_score += score
            except:
                score = 0

            self.results.append((category, result))
            return result

        except Exception as e:
            return f"API Error: {str(e)}"

    def final_report(self):
        avg = self.total_score / len(self.results) if self.results else 0

        report = "\n" + "=" * 60 + "\n"
        report += "GENAI WEEKLY REVIEW REPORT\n"
        report += "=" * 60 + "\n"

        for category, result in self.results:
            report += f"\n[{category}]\n{result}\n"

        report += "\n" + "=" * 60 + "\n"
        report += f"Final Average Score: {avg:.2f}/10\n"

        if avg >= 8:
            report += "Performance Level: Excellent\n"
        elif avg >= 6:
            report += "Performance Level: Good\n"
        else:
            report += "Performance Level: Needs Improvement\n"

        return report


def main():
    print("=" * 60)
    print("ABTalks Season-3 | Task Day-36 | GenAI Weekly Evaluation")
    print("=" * 60)

    evaluator = GenAIEvaluator()

    for item in weekly_review_questions:
        print(f"\nCategory: {item['category']}")
        print(f"Question: {item['question']}")

        answer = input("Your Answer: ")

        result = evaluator.evaluate_response(item["category"], answer)

        print("\nEvaluation Result:")
        print(result)

    print(evaluator.final_report())


if __name__ == "__main__":
    main()
```
